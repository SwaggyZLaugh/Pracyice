```python
import urllib   #获取网页数据

head ={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36"}
```

```python
import datetime

start = '20200116'
end = '20211222'

datestart = datetime.datetime.strptime(start,'%Y%m%d')
dateend = datetime.datetime.strptime(end,'%Y%m%d')
```

```python
date = []
while datestart <= dateend:
    date.append(datestart.strftime("%Y%m%d"))
    datestart += datetime.timedelta(days=1)
```

```python
province_index = {'Beijing':0,'Hubei':16,'Guangdong':18,'Yunnan':24}
```

```python
import pandas as pd
df = pd.DataFrame(index=province_index.keys(),columns=date)
df
```

```python
import json

for i in date:
    url = "http://49.4.25.117/JKZX/yq_{}.json".format(i)
    request = urllib.request.Request(url,headers=head)                #发送请求
    try:
        response = urllib.request.urlopen(request,timeout=5)          #获取反应
        html=response.read().decode("utf-8")                          #获取网页内容
    except urllib.error.URLError as e:
        if hasattr(e,"code"):
              print(e.code)
        if hasattr(e,"reason"):
              print(e.reason)
    data = json.loads(html)
    for j in province_index:
        df[i][j] = data["features"][province_index[j]]["properties"]['新增确诊']
```

```python
df['all'] = [df.sum(axis=1)[0],df.sum(axis=1)[1],df.sum(axis=1)[2],df.sum(axis=1)[3]]
```
